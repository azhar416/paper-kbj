## Encode

### StartTime

```python
all_sensor['StartTime'] = pd.to_datetime(all_sensor['StartTime'], errors='coerce')

all_sensor['StartTimeHour'] = all_sensor['StartTime'].dt.hour
all_sensor['StartTimeMinute'] = all_sensor['StartTime'].dt.minute
all_sensor['StartTimeSecond'] = all_sensor['StartTime'].dt.second
all_sensor = all_sensor.drop(columns=['StartTime'])
```

### SrcAddr & DstAddr & Sport & Dport

```python
# Frequency Encoding
def frequency_encoding(df):
    freq_map = df.value_counts().to_dict()  # Create frequency map
    df = df.map(freq_map)  # Apply frequency encoding
    return df
```

```python
all_sensor['SrcAddr'] = frequency_encoding(all_sensor['SrcAddr'])
all_sensor['DstAddr'] = frequency_encoding(all_sensor['DstAddr'])
all_sensor['Sport'] = frequency_encoding(all_sensor['Sport'])
all_sensor['Dport'] = frequency_encoding(all_sensor['Dport'])
```

### Dir & State & dTos & sTos & Proto

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
all_sensor['Dir'] = le.fit_transform(all_sensor['Dir'].astype(str))
all_sensor['State'] = le.fit_transform(all_sensor['State'].astype(str))
all_sensor['sTos'] = le.fit_transform(all_sensor['sTos'].astype(str))
all_sensor['dTos'] = le.fit_transform(all_sensor['dTos'].astype(str))
all_sensor['BotnetName'] = le.fit_transform(all_sensor['BotnetName'].astype(str))
all_sensor['Proto'] = le.fit_transform(all_sensor['Proto'].astype(str))
```

### Add Target

```python
all_sensor['isBotnet'] = all_sensor['Label'].apply(lambda x: 1 if x == 1 or x == 2 else 0)
all_sensor['isSpam'] = all_sensor['Label'].apply(lambda x: 1 if x == 2 else 0)
```

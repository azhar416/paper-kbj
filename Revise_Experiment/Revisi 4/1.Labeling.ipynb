# Labeling

```python
def categorize_label(label):
    label = label.lower()  
    if 'botnet' in label and 'spam' in label:
        return 2
    elif 'botnet' in label:
        return 1
    else:
        return 0
```

# Sensor 3

```python
df = pd.read_csv(sensor3_path + 'sensor3.binetflow')
```

```python
df['Label'] = df['Label'].apply(categorize_label)
```

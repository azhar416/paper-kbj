# Split Data Train n Test

```python
def split_data(df):
    train, test = train_test_split(df, test_size=0.25, random_state=42)
    return train, test
```

# Sensor 1

```python
df_1 = pd.read_csv(sensor1_path + 'sensor1.binetflow')
```

```python
sensor1_train, sensor1_test = split_data(df_1)
```
